```python
fig = plt.figure(figsize=(16,3))
fig.suptitle("Hubhub Stress Test -- how many requests per second can we handle???", fontsize=16, va='bottom', y=1.0)

def plot_results(results):
  mu = np.mean(results)
  med = np.median(results)
  plt.hist(results, bins=np.linspace(0,8,40))
  plt.axvline(med, color='black', ls='--', label=f"Median ({med:.1f}s)")
  plt.axvline(1, color='red', ls='--', label="Upper limit (1s)")
  plt.xticks(range(9))
  plt.xlabel("Time to handle batch (s)", fontsize=12)
  plt.ylim(0,70)
  plt.legend(fontsize=11)

plt.subplot(151)
plt.title("10 requests per second")
plot_results(results10)
plt.ylabel("Frequency over\n100 repeated batches", fontsize=14)

plt.subplot(152)
plt.title("15 requests per second")
plot_results(results15)

plt.subplot(153)
plt.title("20 requests per second")
plot_results(results20)

plt.subplot(154)
plt.title("25 requests per second")
plot_results(results25)

plt.subplot(155)
plt.title("25 requests per second, Puma")
plot_results(results25_puma)

plt.tight_layout()
plt.show()
```
